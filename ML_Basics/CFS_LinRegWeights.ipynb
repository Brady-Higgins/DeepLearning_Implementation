```python
#Closed Form Solution for weights in linear regression
```

```python
#pandas for data processing
import pandas as pd
#matplotlib for visualizatin
import matplotlib.pyplot as plt
#sklearn for the dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
data = pd.DataFrame(housing.data)
#set columns//features to be equal to the housing.feature names
data.columns = housing.feature_names
data['PRICE'] = housing.target
# remove every feature besides MedInc and PRICE
data = data[['MedInc', 'PRICE']]
data.head()
```

```python
X = data['MedInc'].values
```

```python
import numpy as np
X.shape # (20640,) i.g. a 1D array

design_vector = data['MedInc'].values.reshape(-1,1)       # -> (20640,1) i.g. a column vector
target_vector = data['PRICE'].values.reshape(-1,1)       
```

```python
design_matrix = np.c_[np.ones((X.shape[0], 1)), X]  # Add a bias term (column of ones)

```

Formula for optimizing weights
$$ \mathbf{w} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y} $$


```python
weights = np.linalg.inv(design_matrix.T @ design_matrix) @ design_matrix.T @ target_vector
```

```python
plt.scatter(X,target_vector, color='gray',label='Data points')
plt.plot(X, design_matrix @ weights, color='red', label='Regression line')
```
